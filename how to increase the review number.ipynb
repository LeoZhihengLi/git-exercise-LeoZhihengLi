For example, if the price is too high, the customer may not want to book that Airbnb, and the review number would be small. Variable neighbourhood_group is the area in which Airbnb locates. Various type of Airbnb may engage different people and therefore affect the review number. Variable minimum_nights is the minimum nights that cumstomers need to book for one time or one trip in the Airbnb. Some Airbnb only require an one night minimum booking while others would like to require more than one, such as four nights. For customers who want to stay a night, those Airbnb that reqire more than one night minimum booking may have less review number. Variable availability_365 is the number of days when listing is available for booking during 365 days. It may affect the number of review since in case there are less days avaiable, there may be less number of check in and review.
